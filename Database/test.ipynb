```python
import psycopg2, pandas as pd

conn = psycopg2.connect(
    host="localhost", port=5432,
    dbname="aer_warehouse",
    user="myuser", password="mypass",
)

df = pd.read_sql("""
    SELECT *
    FROM raw.well_gas_analysis__separator_liquid_data
    ORDER BY uwi_short
    LIMIT 200
""", conn)

df.head(100)
```

```python
from sqlalchemy import create_engine, inspect

engine = create_engine("postgresql+psycopg2://myuser:mypass@localhost:5432/aer_warehouse")
insp = inspect(engine)

# All schemas
schemas = insp.get_schema_names()
schemas

# Tables per schema (e.g., raw, curated, dim, public)
for sch in ("raw", "curated", "dim", "public"):
    if sch in schemas:
        print(sch, "->", insp.get_table_names(schema=sch))

# Views per schema (optional)
for sch in ("raw", "curated", "dim", "public"):
    if sch in schemas:
        print(sch, "views ->", insp.get_view_names(schema=sch))
```
